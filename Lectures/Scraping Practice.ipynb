```python
def prettyTime(tm):
	fmt = "%b %d %H:%M"
	datetime_obj_naive = datetime.strptime(tm, "%Y-%m-%dT%H:%M:%S.%fZ")
	return str(datetime_obj_naive.strftime(fmt))
```

```python
def prettyHubSingle(load):
    org = load["origin"]
    des = load["destination"]
    a = ""
    if load["weight"]!="":
        a= "\nWeight: "+load["weight"] 
       
    txt = load["id"]+"\n"+org["city"]+", "+ org["state"]+" ("+org["loading_type"]+") - "+des["city"]+", "+ des["state"]+" ("+des["loading_type"]+")\n"+prettyTime(org["from_date_time"])+" - "+prettyTime(des["from_date_time"])+"\n"+"Miles: "+load["miles"]+a
    
    return txt
```

```python
def getgeodata(cityState):
    cityState = cityState.upper().split(", ")
    res = dfallloc[["lon","lat"]][(dfallloc["city"] == cityState[0]) & (dfallloc["state"] == cityState[1])]
    try:
        if res.empty or not int(cityState[2]):
            return False
    except:
        return False
    return [[res.lon.item(), res.lat.item()], int(cityState[2])]
print(getgeodata("Hebron, KY, 1"))
```
